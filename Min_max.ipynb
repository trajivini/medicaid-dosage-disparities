```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
import us
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
```

```python
df_2019 = pd.read_csv('/Volumes/Data_files/Data_set/SDUD2019.csv')
df_2020 = pd.read_csv('/Volumes/Data_files/Data_set/SDUD2020.csv')
df_2021 = pd.read_csv('/Volumes/Data_files/Data_set/SDUD2021.csv')
df_2022 = pd.read_csv('/Volumes/Data_files/Data_set/SDUD2022.csv')
df_2023 = pd.read_csv('/Volumes/Data_files/Data_set/SDUD2023.csv')
df_2024 = pd.read_csv('/Volumes/Data_files/Data_set/SDUD2024.csv')

index_df = pd.read_csv('/Volumes/Data_files/Data_set/index.csv')
state_df = pd.read_csv('/Volumes/Data_files/Data_set/States_by_Region.csv')
```

```python
df_2019
```

```python
data = pd.concat([df_2019, df_2020, df_2021, df_2022, df_2023, df_2024])
data
```

```python

# data = df_2024
data = data[(data['State'] != 'XX') & (data['State'] != 'AL') & (data['State'] != 'DC') & (data['State'] != 'PR')]
data = data[data['Utilization Type'] == 'MCOU']
data = data[data['Units Reimbursed'] < 317249]
data = data[data['Number of Prescriptions'] < 790]
data = data[data['Total Amount Reimbursed'] < 169373.242]
data = data[data['Medicaid Amount Reimbursed'] < 21542.44]
data = data[data['Non Medicaid Amount Reimbursed'] < 7550.8]
data
```

```python

data

data = data.groupby(['NDC', 'Product Name', 'State', 'Year'])[['Total Amount Reimbursed', 'Units Reimbursed', 'Number of Prescriptions', 'Medicaid Amount Reimbursed', 'Non Medicaid Amount Reimbursed']].sum().reset_index().sort_values('Medicaid Amount Reimbursed', ascending = False)

data
```

```python
data.sort_values('Total Amount Reimbursed')
data = data[
    (data['Units Reimbursed'] != 0) &
    (data['Number of Prescriptions'] != 0) &
    (data['Medicaid Amount Reimbursed'] != 0) &
    (data['Non Medicaid Amount Reimbursed'] != 0)
]

data
```

```python
data = pd.merge(data, index_df, on=['Year', 'State'], how='left')
data = pd.merge(data, state_df, on = 'State', how = 'left')
data = data.drop(['State', 'State f'], axis=1)
data
```

```python
data = data.groupby(['NDC', 'Product Name', 'Region', 'Year']).agg({
    'Total Amount Reimbursed': 'sum',
    'Units Reimbursed': 'sum',
    'Number of Prescriptions': 'sum',  # if you really want sum, not mean
    'Medicaid Amount Reimbursed': 'sum',
    'Non Medicaid Amount Reimbursed': 'sum',
    'Temp Index': 'mean',
    'poverty index': 'mean',
    'GDP per state': 'mean',
    'personal income': 'mean',
    'population estimates': 'mean'
}).reset_index().sort_values('Medicaid Amount Reimbursed', ascending=False)
data
```

```python
scaling = MinMaxScaler()
scaled_array = scaling.fit_transform(data[['Units Reimbursed']])
scaled_df = pd.DataFrame(scaled_array, columns=['Units Reimbursed (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['Total Amount Reimbursed']])
scaled_df = pd.DataFrame(scaled_array, columns=['Total Amount Reimbursed (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['Number of Prescriptions']])
scaled_df = pd.DataFrame(scaled_array, columns=['Number of Prescriptions (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['Medicaid Amount Reimbursed']])
scaled_df = pd.DataFrame(scaled_array, columns=['Medicaid Amount Reimburse (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['Non Medicaid Amount Reimbursed']])
scaled_df = pd.DataFrame(scaled_array, columns=['Non Medicaid Amount Reimbursed (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['Temp Index']])
scaled_df = pd.DataFrame(scaled_array, columns=['Temp Index (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['poverty index']])
scaled_df = pd.DataFrame(scaled_array, columns=['poverty index (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['GDP per state']])
scaled_df = pd.DataFrame(scaled_array, columns=['GDP per state (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['personal income']])
scaled_df = pd.DataFrame(scaled_array, columns=['personal income (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

scaled_array = scaling.fit_transform(data[['population estimates']])
scaled_df = pd.DataFrame(scaled_array, columns=['population estimates (Scaled)'])
data = pd.concat([data.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)





data
```

```python
def target_mean_encoding(df, categorical_col, target_col, n_splits = 5):
    df_encoded = pd.Series(np.zeros(df.shape[0]), index = df.index)
    kf = KFold(n_splits = n_splits, shuffle = True, random_state = 4)

    for train_index, val_index in kf.split(df):
        train_data, val_data = df.iloc[train_index], df.iloc[val_index]
        target_mean = train_data.groupby(categorical_col)[target_col].mean()
        df_encoded.iloc[val_index] = val_data[categorical_col].map(target_mean)

    global_mean = df[target_col].mean()
    df_encoded.fillna(global_mean, inplace = True)
    return df_encoded
```
